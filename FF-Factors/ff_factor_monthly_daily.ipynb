```python
import wrds
import pandas as pd

# Connect to WRDS
db = wrds.Connection()

# === Retrieve monthly Fama-French factors (3 factors + momentum) ===
ff_monthly = db.get_table(library='ff_all', table='factors_monthly')
ff_monthly['date'] = pd.to_datetime(ff_monthly['date'])

# Filter for 2020-11 to 2021-04
ff_monthly_filtered = ff_monthly[(ff_monthly['date'] >= '2020-11-01') & (ff_monthly['date'] <= '2021-04-30')]

# Save to CSV
ff_monthly_filtered.to_csv('ff_factors_monthly_202011_202104.csv', index=False)


# === Retrieve daily Fama-French factors (3 factors + momentum) ===
ff_daily = db.get_table(library='ff_all', table='factors_daily')
ff_daily['date'] = pd.to_datetime(ff_daily['date'])

# Filter for 2020-11 to 2021-04
ff_daily_filtered = ff_daily[(ff_daily['date'] >= '2020-11-01') & (ff_daily['date'] <= '2021-04-30')]

# Save to CSV
ff_daily_filtered.to_csv('ff_factors_daily_202011_202104.csv', index=False)
```
